### Prototipar query

**staging.dim_preco**

```python
(df_gestao_faltas[['PROJETO','PN_MAT_FALT']].drop_duplicates()
 .merge(df_fact_top_material[['FTOP_PART_NUMBER','FTOP_USD_PRECO_UNIT']].drop_duplicates(),
       left_on='PN_MAT_FALT',
       right_on='FTOP_PART_NUMBER')
 .drop(columns=['FTOP_PART_NUMBER'])
)
```

```python
query = \
"""
with
    ftm as (
        SELECT DISTINCT
            FTOP_PART_NUMBER
            , FTOP_USD_PRECO_UNIT
        FROM
            raw.FACT_TOP_MATERIAL
),
    gf as (
        SELECT DISTINCT
            PROJETO
            , PN_MAT_FALT
        FROM
            raw.GESTAO_FALTAS
)

select
    gf.PROJETO
    , gf.PN_MAT_FALT
    , ftm.FTOP_USD_PRECO_UNIT
from gf
join ftm
    on gf.PN_MAT_FALT = ftm.FTOP_PART_NUMBER
"""

with duckdb.connect(str(duckdb_file)) as con:    
    result_ = con.execute(query).fetchdf()
result_
```

```python
query = \
"""
select * from dbt_STAGING.dim_preco
"""

with duckdb.connect(str(duckdb_file)) as con:    
    df_dim_preco = con.execute(query).fetchdf()
df_dim_preco
```

**serving.dim_projeto_stats**

contagem de material material faltante e preço médio por projeto

```python
(df_dim_preco
 .groupby("PROJETO").agg({"PN_MAT_FALT": "count", 
                          "FTOP_USD_PRECO_UNIT": "mean"})
 .reset_index()
 .sort_values('PROJETO')
)
```

```python
query = \
"""
select
    PROJETO
    , count(PN_MAT_FALT) as "COUNT_PN_MAT_FALT"
    , avg(FTOP_USD_PRECO_UNIT) as "AVG_FTOP_USD_PRECO_UNIT"
from dbt_STAGING.dim_preco
group by *
order by "PROJETO"
"""

with duckdb.connect(str(duckdb_file)) as con:    
    df_dim_projeto_stats = con.execute(query).fetchdf()
df_dim_projeto_stats
```
